```python
'''
# Import / Read Dataset
df = pd.read_excel(r"/content/Python Dynamic DataSet.xlsx")

# Display Dataset
df.head()

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go

# Create a Dash application
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Financial Analysis Dashboard", style={'text-align': 'center'}),


    # Dropdown to select a plot
    dcc.Dropdown(
        id='plot-selector',
        options=[
            {'label': 'Countrywise Revenue', 'value': 'revenue_histogram'},
            {'label': 'Region Pie Chart', 'value': 'region_pie'},
            {'label': 'Quarterwise Revenue Bar Chart', 'value': 'quarter_bar'},
            {'label': 'Month and Revenue Line Plot', 'value': 'quarter_line'},
            {'label': 'Revenue Scatter Plot', 'value': 'revenue_scatter'},
        ],
        value='revenue_histogram'
    ),

    # Placeholder for the selected plot
    dcc.Graph(id='selected-plot'),
])

# Define callback to update the selected plot based on user's choice
@app.callback(
    Output('selected-plot', 'figure'),
    [Input('plot-selector', 'value')]
)
def update_selected_plot(selected_value):
    if selected_value == 'revenue_histogram':
        # Countrywise Revenue Histogram
        fig = px.histogram(df, x='Country', title='Country', y='Revenue (USD)', labels={'Revenue (USD)': 'Revenue (USD)'})
        fig.update_traces(marker_color=px.colors.sequential.Viridis, opacity=0.7)
        return fig
    elif selected_value == 'region_pie':
        # Group by Region and sum the Revenue
        region_pie = df.groupby('Region')['Revenue (USD)'].sum().reset_index()

        # Region Pie Chart
        fig = go.Figure(data=[go.Pie(labels=region_pie['Region'], values=region_pie['Revenue (USD)'])])
        fig.update_layout(title='Revenue Distribution by Region')
        return fig
    elif selected_value == 'quarter_bar':
        # Rank and Revenue Histogram
        fig = px.histogram(df, x='Quarter', title='Quarter', y='Revenue (USD)', labels={'Quarter': 'Quarter', 'Revenue (USD)': 'Revenue (USD)'})
        fig.update_traces(marker_color=px.colors.sequential.Viridis, opacity=0.7)
        return fig
    elif selected_value == 'quarter_line':
        # Month and revenue Line Plot
        fig = px.line(df, x='Month', y='Revenue (USD)', title='Revenue (USD)', labels={'Revenue (USD)': 'Revenue (USD)'})
        return fig
    elif selected_value == 'revenue_scatter':
        # Revenue Scatter Plot
        fig = px.scatter(df, y='Revenue (USD)', color='Products', title='Revenue', labels={'Revenue (USD)': 'Revenue (USD)', 'Products': 'Products'})
        return fig

# Run the app on a different port
if __name__ == '__main__':
    app.run_server(debug=True, port=8051)

    '''

```
